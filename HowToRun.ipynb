## How to run
### Prerequisite

- Make sure you have aws account and you are in ap-south-1 Region (Mumbai)
- Check and validate the AMI with instance type in ap-south-1 and match in file://stack-003.yml
    Line number- 321, 322
```
          ap-south-1:
          HVM64: ami-05ba3a39a75be1ec4  
```
- You could edit above as per your region.

### Run your code

- Place the file in aws > CloudFormation > Create Stack


- Give
```
 stack name: RStudio-connect-Installation-01
 Parameters-
      Instance Type: t2.micro
      keyName: Of your choice
```
